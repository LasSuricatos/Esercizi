##### <span style="color:red"> Il dataframe </span>
Il dataset si riferisce ai matrimoni celebrati a Milano per anno dell'evento, dal 2003 in poi.
<p>In particolare questo dataframe è disponibile sul sito del comune di Milano, i dati raccolti al suo interno sono stati collezionati a partire dal primo gennaio del 2003 fino al 31 dicembre 2019 e sono stati aggiornati annualmente. </p>

Cliccare qui andare al sito degli [Open data del comune di Milano](https://dati.comune.milano.it/).


###### <span style="color:red"> Le variabili </span>
I matrimoni sono classificati secondo variabili relative all'evento (luogo del matrimonio, tipo di rito), e variabili relative ai due sposi (residenza, stato civile precedente il matrimonio, cittadinanza, età).
<p> La variabile 'numerosità' si riferisce a quante persone si sono sposate con le stesse caratteristiche. </p>

```python
#importo le librerie
import numpy as np
import pandas as pd

#importo il file csv
data =pd.read_csv(r"C:\Users\Laura\Desktop\PythonDataScienceHandbook-master\notebooks\data\matrimoni_unioni_civili.csv")

#cancello la colonna '_id' in quanto jupyter notebook al momento dell'importazione del file csv ne crea una identica
data.drop('_id', inplace=True, axis=1)

#visualizzo il dataframe
data
```

###### <span style="color:red"> L'analisi dei dati </span>
<p> Creiamo due mask di valori booleani rispetto alle condizioni se il rito è cattolico o civile, dopo di chè andremo a confrontare con delle misure statistiche quali media,mediana,massimo e minimo le persone che si sono sposate con rito cattolico con quelle che si sono sposate con rito civile nell'anno 2019 </p>


```python
mask_cattolico= data['Rito'] == 'Cattolico'
```

```python
mask_civile= data['Rito'] == 'Civile'
```

```python
print(" DATAFRAME DELLE PERSONE SPOSATE CON RITO CATTOLICO NEL 2019 \n")
cattolici2019 = data.loc[(data['Rito'] == 'Cattolico') & (data['Anno_evento'] == 2019)]
cattolici2019
```

```python
print("STATISTICHE RELATIVE ALLE PERSONE SPOSATE CON RITO CATTOLICO NEL 2019 \n")
print("Media delle persone che si sono sposate con il rito cattolico: ", np.mean(cattolici2019['Numerosità']))
print("Mediana delle persone che si sono sposate con il rito cattolico: ", np.median(cattolici2019['Numerosità']))
print("Massimo delle persone che si sono sposate con il rito cattolico: ", np.max(cattolici2019['Numerosità']))
print("Minimo delle persone che si sono sposate con il rito cattolico: ", np.min(cattolici2019['Numerosità']))
```

```python
print(" DATAFRAME DELLE PERSONE SPOSATE CON RITO CIVILE NEL 2019 \n")
civili2019 = data.loc[(data['Rito'] == 'Civile') & (data['Anno_evento'] == 2019)]
civili2019
```

```python
print("STATISTICHE RELATIVE ALLE PERSONE SPOSATE CON RITO CIVILE NEL 2019 \n")
print("Media delle persone che si sono sposate con il rito civile: ", np.mean(civili2019['Numerosità']))
print("Mediana delle persone che si sono sposate con il rito civile: ", np.median(civili2019['Numerosità']))
print("Massimo delle persone che si sono sposate con il rito civile: ", np.max(civili2019['Numerosità']))
print("Minimo delle persone che si sono sposate con il rito civile: ", np.min(civili2019['Numerosità']))
```

##### <span style="color:red"> Commento alle statistiche </span>
<p> Nel 2019 possiamo osservare le persone preferiscono sposarsi in comune con rito civile; </p>
<p> Il massimo di persone sposate con rito cattolico con le stesse caratteristiche nel 2019 sono state 7 coppie. </p>
 <p>   Il massimo di persone sposate con rito civile con le stesse caratteristiche nel 2019 sono state 11 coppie.
</p> <p> In base alle statistiche possiamo notare come nel 2019 ci sia una tendenza a sposarsi con rito civile.


#####  <span style="color:red">GRAFICI TENDENZE MATRIMONI 2019</span>


```python
#Importazione librerie
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
print("Scatterplot dell'andamento dei matrimoni dei diversi dal 2003 fino al 2019 diversificati per rito")
seaborn.scatterplot(data=data, x='Anno_evento', y='Numerosità', hue='Rito', style='Rito' );
```

##### <span style="color:red">DATASET DEI MATRIMONI CON RITO CATTOLICO E CIVILE NEL CORSO DEGLI ANNI </span>

```python
cattolici = data.loc[(data['Rito'] == 'Cattolico')]
```

```python
civili = data.loc[(data['Rito'] == 'Civile')]
```

```python
seaborn.histplot(data=cattolici, x='Anno_evento', stat='count', kde='true',color='green')
plt.xlabel('Anno')
plt.ylabel("Numero matrimoni");
#Come abbiamo già notato con il calcolo delle statistiche e la rappresentazione grafica in precedenza i matrimoni 
# con rito cattolico sono diminuiti nel corso degli anni
```

```python
seaborn.histplot(data=civili, x='Anno_evento', stat='count', kde='true',color='purple')
plt.xlabel('Anno')
plt.ylabel("Numero matrimoni");
#Notiamo un andamento non troppo regolare, in particolare possiamo notare come nel 2008 c'è stata un diminuizione 
#fino al 2010, dove sono state date delle delucidazioni dalla Corte Costituzionale sulla legge sulle unioni civili.
```

##### <span style="color:red">Commento dei vari stati civili e della cittadinanza dei partner</span>

```python
seaborn.histplot(data=civili, x='Stato_civile_moglie', stat='probability',color="black")
plt.xticks(rotation='vertical');
seaborn.histplot(data=cattolici, x='Stato_civile_moglie', stat='probability',color="yellow")
plt.xticks(rotation='vertical')
plt.xlabel('Stato civile moglie');
```

<p> Circa il 95% delle donne nubili si sposano con rito cattolico (colonna giallo) mentre le donne nubili che si sposano con il rito civile sono circa il 78% </p>
<p> Le donne che divorziano e che si risposano con rito civile sono circa il 20%, le donne che si risposano con rito cattolico sono poche in quanto la procedura per poterlo fare è complessa e lunga. </p>

```python
seaborn.histplot(data=civili, x='Stato_civile_marito', stat='probability',color="blue")
seaborn.histplot(data=cattolici, x='Stato_civile_marito', stat='probability',color="pink")
plt.xticks(rotation='vertical')
plt.xlabel('Stato civile marito');
```

<p> Nei riti civili, cosi come anche riti cattolici, la maggior parte degli uomini che si sposano sono celibi </p>
<p> In particolare i celibi che si sposano con rito cattolico sono quasi il 100% mentre i celibi che si sposano con rito civile sono circa il 70% </p>
<p>Gli uomini divorziati che si risposano con rito civile sono circa il 23% mentre per quanto riguardo il rito cattolico sono in numero esiguo per le stesse motivazioni date precendentemente </p>


```python
seaborn.histplot(data=civili, x='Residenza_moglie', stat='probability',color="grey")
seaborn.histplot(data=cattolici, x='Residenza_moglie', stat='probability',color="lightblue",alpha=0.5)
plt.xlabel('Cittadinanza moglie');
```

<p> Naturalmente la maggior parte delle donne che si sposano sono residenti a Milano in quanto i dati sono stati raccolti a Milano </p>
<p> Le donne che si sono sposate con rito cattolico con residenza diversa da Milano sono circa il 28% mentre le donne sposate con rito civile sono circa l'8% </p>
<p> Le donne che hanno una residenza estera e che si sono sposate a Milano con rito cattolico sono circa il 3% mentre quelle sposate con rito civile sono quasi il 20%</p>


```python
seaborn.histplot(data=civili, x='Residenza_marito', stat='probability',color="orange")
seaborn.histplot(data=cattolici, x='Residenza_marito', stat='probability',color="brown")
seaborn.set_theme(style='darkgrid', palette='deep')
plt.xlabel('Residenza marito');
```

<p> Gli uomoni che si sposano con rito cattolico con residenza in Italia, ma non a Milano, sono quasi il 40% mentre quelli con rito civile sono il 10%</p>
<p> Gli uomini che si sposano con rito cattolico con residenza estera sono circa il 3% mentre coloro che si sposano con rito civile e residenza estera sono circa il 12% <p>
