### Esercizio: What is the Average Height of US Presidents?

```python
import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\Users\Laura\Desktop\PythonDataScienceHandbook-master\notebooks\data\president_heights.csv")
print("DATAFRAME UTILIZZATO \n",data)
```

```python
#per far si che np sia definito biosgna importare numpy as np
#la funzione np.array(data[...]) crea un array con gli elementi indicati nelle parentesi quadre e con il dataframe indicato (data nel mio caso)
heights= np.array(data['height(cm)'])
print("ARRAY DELLE ALTEZZE \n",heights)
names= np.array(data['name'])
print("\nARRAY DEI NOMI \n", names)
```

###### INFORMAZIONI STATISTICHE
* La media è il risultato dell'addizione dei valori tra di loro e della loro divisione per il numero dei dati
* la mediana è il valore che occupa la posizione centrale
* la deviziaione standard è una stima della variabilità della popolazione di dati 
* il 25esimo percentile è il primo quartile
* il 75esimo percentile è il terzo quartile

```python
print("La media delle altezze è di: ", heights.mean())
print("La deviazione standard delle altezze è pari a: ", heights.std())
print("Altezza minima: ", heights.min())
print("Altezza massima: ", heights.max())
print("25esimo percentile delle altezze: ", np.percentile(heights, 25))
print("Mediana delle altezze: ", np.median(heights))
print("75esimo percentile delle altezze: ", np.percentile(heights, 75))
```

###### ISTOGRAMMA DELLE ALTEZZE

```python
#Importazione librerie
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
```

```python
#la funzione hist crea un istogramma con l'elemento dato
#la funzione title attribuisce un titolo al grafico
#le funzioni xlabel e ylabel creano delle etichette per le variabili sull'asse delle x e delle y
plt.hist(heights)
plt.title('Distribuzione delle altezze dei Presidenti degli USA')
plt.xlabel('altezza (cm)')
plt.ylabel('numero');
seaborn.set_theme(font='sans-serif',style='darkgrid')
```

```python
print("Variazione del numero di colonne e range massimo e minimo per l'asse x\n")
plt.hist(heights,bins=20,range=[100,200])
plt.title('Distribuzione delle altezze dei Presidenti degli USA')
plt.xlabel('altezza (cm)')
plt.ylabel('numero');
```

```python
print("DIVERSO TIPO DI ISTOGRAMMA \n")
plt.hist(heights, histtype='step',orientation='horizontal')
plt.title('Distribuzione delle altezze dei Presidenti degli USA')
plt.xlabel('altezza (cm)')
plt.ylabel('numero');
```

```python
print("ISTOGRAMMA CON FUNZIONE SEABORN \n")
seaborn.histplot(data=heights,binwidth=4,stat='frequency',color='red',kde='true',element='step');
```
