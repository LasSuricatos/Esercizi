```python
#Esercizio presidenti US 14/10
import numpy as np
import pandas as pd
data=pd.read_csv(r"C:\Users\utente\Desktop\PythonDataScienceHandbook-master\notebooks\data\president_heights.csv")
print("dataframe utlizzato:\n",data)
```

```python
#Estrarre primi dieci e media altezze 
primi_dieci=data[:10]
print("primi dieci presidenti:\n",data[:10])
altezza_primi_dieci=np.array(primi_dieci["height(cm)"])
print("\n array altezze primi dieci:\n",altezza_primi_dieci)
print("\n media altezza primi dieci:\n",altezza_primi_dieci.mean())
```

```python
#estrarre ultimi dieci e fare media altezze 
ultimi_dieci=data[32:]
print("ultimi dieci presidenti:\n",data[32:])
altezza_ultimi_dieci=np.array(ultimi_dieci["height(cm)"])
print("\n array altezza ultimi dieci:\n",altezza_ultimi_dieci)
print("\n media altezza ultimi dieci:\n",altezza_ultimi_dieci.mean())
```

```python
#estrarre i presidenti di altezza compresa tra 170 e 180
altezza170180=data.loc[(data["height(cm)"]>=170) & (data["height(cm)"]<=180)]
print("presidenti tra 170 e 180:\n",altezza170180)
```
